```julia
# Sierpinski
function T(x,p)
#     p = pop!(rand(1))
    if p < 1/3
#         println("1")
        y = [[0.5 0];[0 0.5]]*x
    elseif 1/3 <= p <= 2/3
#         println("2")
        y = [[0.5 0];[0 0.5]]*x + [0.5;0]
    else
#         println("3")
        y = [[0.5 0];[0 0.5]]*x + [0.25;sqrt(3)/4]
    end
    return y
end

N = 10000
x0 = [0;0]
x = []
push!(x,x0)
p = rand(N)
for i = 1:N
    y = x[i]
    y = T(y,pop!(p))
    push!(x,y)
end
```

```julia
scatter(map(x->x[1],x),map(x->x[2],x),markersize=1,c=:black)
```
